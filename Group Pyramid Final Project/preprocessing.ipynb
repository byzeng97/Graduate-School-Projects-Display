```python
import pandas as pd

df = pd.read_csv("YTS.csv")
df = df.rename(columns={"LocationAbbr": "STATE"})
```

```python
# filter year from 2014 to 2017
df = df[(df.YEAR>=2014)&(df.YEAR<=2017)]
df.dropna(inplace=True)
df = df.reset_index(drop=True)
```

```python
# split Geolocation into lat and lng
df['GeoLocation'] = df['GeoLocation'].apply(lambda x: x[1:-1])
df['lat'] = df['GeoLocation'].apply(lambda x: x.split(',')[0])
df['lng'] = df['GeoLocation'].apply(lambda x: x.split(',')[1])

# Drop Geo
df.drop(['GeoLocation'], axis=1, inplace=True)
```

```python
# load secondary data source
heart = pd.read_csv("heart.csv")
overdose = pd.read_csv("overdose.csv")
stroke = pd.read_csv("stroke.csv")
cancer = pd.read_csv("cancer.csv")
```

```python
# merge the data
df = df.merge(heart, how='left', on=['YEAR', 'STATE'])
df = df.merge(overdose, how='left', on=['YEAR', 'STATE'])
df = df.merge(stroke, how='left', on=['YEAR', 'STATE'])
df = df.merge(cancer, how='left', on=['YEAR', 'STATE'])
```

```python
df = df.dropna()
```

```python
df.to_csv("cleaned_YTS", index=False)
```
