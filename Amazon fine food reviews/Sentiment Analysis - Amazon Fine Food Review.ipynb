```python
import pandas as pd
import nltk
import random
nltk.download('vader_lexicon')
```

```python
df=pd.read_csv('Reviews.csv')
df
```

```python
df['Text']
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
df1 = df.sample(n=20000,frac=None, replace=False, weights=None, random_state=1, axis=None)
```

```python
df1['Text']
```

```python
sid = SentimentIntensityAnalyzer()
for sentence in df1['Text']:
    ss = sid.polarity_scores(sentence)
    for k in ss:
         print('{0}: {1},  '.format(k, ss[k]), end='')
    print()
```

```python

review_neg = []
review_neu = []
review_pos = []
```

```python
df1['Text'].iloc[0]
```

```python
sid = SentimentIntensityAnalyzer()
for i in range(len(df1['Text'])):
    ss = sid.polarity_scores(df1['Text'].iloc[i])
    review_neg.append(ss["neg"])
    review_neu.append(ss["neu"])
    review_pos.append(ss["pos"])
```

```python
data = {'review_neg': review_neg, 'review_neu': review_neu, 'review_pos': review_pos}
dataframe = pd.DataFrame(data=data)
dataframe
```

```python
dataframe.review_pos.mean()
```

```python
dataframe.review_neg.mean()
```

```python
df1['review_pos'] = review_pos
df1['review_neu'] = review_neu
df1['review_neg'] = review_neg
df1
```

```python
df1.to_csv('Sentiment_Review.csv')
```
