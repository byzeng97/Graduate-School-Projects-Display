```python
from google.cloud import storage
import os
from io import BytesIO
import pandas as pd

#Connect to GCP bucket and assign the bucket_name and specify the file name
bucket_name = "assignment3-budget" #Assign the bucket name where your file is stored
storage_client = storage.Client()

bucket = storage_client.get_bucket(bucket_name)
```

```python
blob = storage.blob.Blob("Reviews.csv",bucket)
blob
```

```python
# Convert to a pandas dataframe
content = blob.download_as_string()
train = pd.read_csv(BytesIO(content))
```

```python
train
```

```python
# put the userid to a set
userid_set = set(train['UserId'])
productid_set = set(train['ProductId'])
```

```python
userid_list = list(userid_set)
productid_list = list(productid_set)
```

```python
# assign numeric id to userid
userid_assign = []
for i in range(len(userid_list)):
    i += 1
    userid_assign.append(i)
userid_assign
```

```python
# assign numeric id to productid
productid_assign = []
start_num = 300000
for k in range(len(productid_list)):
    k = k + start_num
    productid_assign.append(k)
productid_assign
```

```python
userid_df = pd.DataFrame({'userid_new': userid_assign,
                             'userid_old': userid_list})
userid_df
```

```python
productid_df = pd.DataFrame({'productid_new': productid_assign,
                            'productid_old': productid_list})
productid_df
```

```python
productid_df['productid_new'][1]
```

```python
# join dataframe with numeric productid to the raw dataset
kk = train.set_index('ProductId').join(productid_df.set_index('productid_old'))
```

```python
# join dataftame with numeric userid to the raw dataset
new_data = kk.set_index('UserId').join(userid_df.set_index('userid_old'))
new_data
```

```python
recommendation_train = new_data[['userid_new','productid_new','Score']]
```

```python
# impport everything we need for the rec sys
from pyspark.sql.types import *
from pyspark import SparkContext
from pyspark.sql import SQLContext
from pyspark.ml.evaluation import RegressionEvaluator
from pyspark.ml.recommendation import ALS
from pyspark.sql import Row
from pyspark.sql import SparkSession
```

```python
# initialize spark session
spark = SparkSession.builder.appName('Recommendation_system').getOrCreate()
```

```python
# create a spark dataframe from a pandas dataframe
df = spark.createDataFrame(recommendation_train)
```

```python
# create testing and training data
(training, test) = df.randomSplit([0.8, 0.2])
```

```python
als = ALS(maxIter=5, regParam=0.01, userCol="userid_new", itemCol="productid_new", ratingCol="Score",coldStartStrategy="drop")
```

```python
model = als.fit(training)
```

```python
# generate predictions and evaluations
predictions = model.transform(test)
```

```python
evaluator = RegressionEvaluator(metricName="rmse", labelCol="Score",predictionCol="prediction")
```

```python
rmse = evaluator.evaluate(predictions)
print("Root-mean-square error = " + str(rmse))
```
